```python
import pandas as pd
import requests

def listar_opcoes(ativo_objeto, vencimento):
    url = f'https://opcoes.net.br/listaopcoes/completa?idAcao={ativo_objeto}&listarVencimentos=false&cotacoes=true&vencimentos={vencimento}'
    r = requests.get(url).json()
    l = [[ativo_objeto, vencimento, i[0].split('_')[0], i[2], i[3], i[4], i[5], i[8]] for i in r['data']['cotacoesOpcoes']]
    return pd.DataFrame(l, columns=['ativo_objeto', 'vencimento', 'ticker', 'tipo', 'modelo', 'O/A/ITM', 'strike', 'preco'])

listar_opcoes('PETR4', '2023-04-20')
```

```python

def listar_tudo(ativo_objeto):
    url = f'https://opcoes.net.br/listaopcoes/completa?idAcao={ativo_objeto}&listarVencimentos=True&cotacoes=true'
    r = requests.get(url).json()
    vencimentos = [i['value'] for i in r['data']['vencimentos']]
    df = pd.concat([listar_opcoes(ativo_objeto, vencimento) for vencimento in vencimentos])
    return df

listar_tudo('PETR4')
```
