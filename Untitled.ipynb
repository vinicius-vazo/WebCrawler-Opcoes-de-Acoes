```python
import pandas as pd
import requests

def listar_opcoes(ativo_objeto):
    url = f'https://opcoes.net.br/listaopcoes/completa?idAcao={ativo_objeto}&listarVencimentos=True&cotacoes=true'
    r = requests.get(url).json()
    l = [[i[0].split('_')[0], i[2], i[3], i[5], i[8]] for i in r['data']['cotacoesOpcoes']]
    return pd.DataFrame(l, columns=['ativo', 'tipo', 'modelo', 'strike', 'preco'])

listar_opcoes('PETR4')
```
